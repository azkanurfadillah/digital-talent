```python
import urllib

r = urllib.request.urlopen('http://tekno.kompas.com/').read()
print(type(r))
print(r[1:1000])
```

```python
from bs4 import BeautifulSoup 

soup = BeautifulSoup(r,"lxml")
print(type(soup))
print(soup.prettify()[1:1000])
```

```python
links = soup.find_all('a')
print(len(links))

for link in links:
    print('%s: %s'%(link.text.strip(),link["href"]))
```

```python
berita_link = [link for link in links if '/read/' in str(link) and 'javascript:void(0)' not in str(link) 
               and '#' not in str(link)]
print(len(berita_link))
for link in berita_link:
    print('%s: %s'%(link.text.strip(),link["href"]))
```

```python
html = urllib.request.urlopen("http://tekno.kompas.com/read/2016/04/26/19500067/6.Aplikasi.Gratis.Pengirit.Baterai.Android").read()
soup = BeautifulSoup(html, "lxml")

judul = soup.find("div","col-bs10-10").find('h1').text.strip()
isi_berita = soup.find("div","read__content").text.strip()
tanggal = soup.find("div","article__date").text

print(tanggal.strip())
print(judul)
tanggal = tanggal.strip().split(',')[1].split('|')[0].strip()
print(tanggal)
print(isi_berita)
```
