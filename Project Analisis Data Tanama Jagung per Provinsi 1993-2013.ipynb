## Produksi tanaman jagung per provinsi tahun 1993-2013

Data ini berisi jumlah produksi(ton) tanaman pangan jagung per provinsi dari tahun 1993- 2013, juga berisi luas panen(hektar) dan produktivitas (kwintal/hektar) yang apabila dikalikan akan menghasilkan angka produksi(ton).

```python
import pandas as pd
tanaman_jagung = pd.read_csv("tanaman-pangan-jagung-per-provinsi-1993-2013.csv", delimiter = ",", encoding = "utf-8")
```

```python
tanaman_jagung
```

```python
tanaman_jagung.columns
```

#### Produksi tanaman jagung tahun 2013 per provinsi

```python
tanaman_jagung[tanaman_jagung['tahun']==2013]
```

### Jumlah produksi tanaman jagung tahun 1993-2013 per provinsi

```python
tanaman_jagung.groupby('nama_provinsi').sum() ['produksi']
```

```python
ax = tanaman_jagung.groupby('nama_provinsi').sum() ['produksi']
ax = ax.plot(kind='bar', title ="Jumlah Produksi Tahun 1993- 2013 Berdasarkan Provinisi", figsize=(20,8), legend=True, fontsize=13)
ax.set_xlabel("nama provinsi", fontsize=12)
ax.set_ylabel("jumlah produksi", fontsize=12)
ax.get_yticks(['produksi']) 
# yticks=('produksi')
#locs, labels = yticks (["produksi"])
```

### Jumlah produksi dari tahun 1993-2013

```python
tanaman_jagung.groupby('tahun').sum() ['produksi']
```

```python
ax = tanaman_jagung.groupby('tahun').sum() ['produksi']
ax = ax.plot(kind='bar', title ="Jumlah Produksi Berdasarkan Tahun", figsize=(20,8), legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Jumlah Produksi", fontsize=12)
ax.get_yticks(['produksi']) 
# yticks=('produksi')
#locs, labels = yticks (["produksi"])
```
